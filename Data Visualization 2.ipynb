```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
df
```

```python
df.isnull().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].mode()[0])
```

```python
sns.heatmap(df.isnull(),cmap='viridis')
```

```python
# Scatter plot to visualize the relationship between survivsl rate and ticket fare
plt.figure(figsize=(8, 8))
sns.scatterplot(x='Survived',y='Fare',data=df,color='Green')
plt.title("Distribution of survived vs Fare")
plt.xlabel('survived')
plt.ylabel('Fare')
plt.show()
```

```python
# Scatter plot to visualize the relationship between age and ticket fare
plt.figure(figsize=(8,4))
sns.scatterplot(x='Age',y='Fare',data=df,color='Green')
plt.title("Distribution of Age vs Fare")
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# Bar plot to visualize survival rates by passenger class
rate_by_class=df.groupby('Pclass')['Survived'].mean()*100
#gives survival rate class wise

plt.figure(figsize=(8, 6))
sns.barplot(x=rate_by_class.index, y=rate_by_class.values)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate (%)')
plt.show()

# Print the survival rates
print(rate_by_class)
```

```python
# Box plot for distribution of age with respect to gender and survival
plt.figure(figsize=(10, 8))
sns.boxplot(x='Sex', y='Age', hue='Survived', data=df, palette='husl')
plt.title('Box Plot: Distribution of Age with Respect to Gender and Survival')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```
