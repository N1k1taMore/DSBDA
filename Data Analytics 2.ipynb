```python
# import pandas as pd
# import numpy as np
# import matplotlib.pyplot as plt
# from sklearn.model_selection import train_test_split
# from sklearn.preprocessing import StandardScaler 
# from sklearn.linear_model import LogisticRegression
# from sklearn.metrics import confusion_matrix, precision_score, accuracy_score, recall_score, f1_score
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, recall_score, precision_score, f1_score, accuracy_score
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
```

```python
#scatterpot to see data distribution
plt.scatter(dataset.Age, dataset.Purchased, marker='*', color='blue')
plt.xlabel('Age')
plt.ylabel('Purchased')
plt.title('ScatterPlot of Age Vs Purchased')
plt.show()
```

```python
#spliting data into independent and dependent variables
X = dataset[['Age', 'EstimatedSalary']]
y = dataset['Purchased']
```

```python
#spliting datset into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.10, random_state=42)
```

```python
#feature scaling to bring variables into single scale
sc = StandardScaler()
print(X_train)
X_train = sc.fit_transform(X_train)
print(X_train)
X_test = sc.fit_transform(X_test)
```

```python
#creating model and fitting logisticregression to training set
logisticRegression = LogisticRegression()
logisticRegression.fit(X_train, y_train)
```

```python
#predict test set results
y_pred = logisticRegression.predict(X_test)
print(y_pred)
```

```python
#actual vs predicted value side by side
y_compare = np.vstack((y_test, y_pred)).T
print("Actual vs Predicted Values:")
print(y_compare)
```

```python

plt.figure(figsize=(10,5))
bar_width = 0.35
index = np.arange(len(y_test))

plt.bar(index,y_test, width=bar_width, label='Actual', color='blue')
plt.bar(index+bar_width,y_pred, width=bar_width, label='Predicted', color='orange')

plt.xlabel('Sample Index')
plt.ylabel('Purchased')
plt.title('Actual vs Predicted Values')
plt.legend()
plt.show()
```

```python
accur = accuracy_score(y_test, y_pred)
print("accuracy score is : " ,accur)
```

```python
# Making the Confusion Matrix
#The .ravel() function in NumPy returns a contiguous flattened array. 
#It collapses the array into a one-dimensional array, flattening it row by row.

cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
TN, FP, FN, TP = cm.ravel()
print('True Negatives (TN): ', TN)
print('True Positives (TP): ', TP)
print('False Positives (FP): ', FP)
print('False Negatives (FN): ', FN)
```

```python
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap="Blues")
plt.title('Confusion Matrix - Actual vs Predicted')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.show()
```

```python
# Error rate
error_rate = (FP + FN) / (TN + TP + FP + FN)
print(error_rate)

```

```python
# Precision
precision = precision_score(y_test, y_pred)
print(precision)

```

```python
# Recall
recall = recall_score(y_test, y_pred)
print(recall)

```

```python
# F1 Score
f1 = f1_score(y_test, y_pred)
print(f1)
```

```python
# Computing Average precision score from predicted scores
from sklearn.metrics import average_precision_score
print('The Average Precision score for logistic clasifier is: ',average_precision_score(y_test, y_pred) )
```
