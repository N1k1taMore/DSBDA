```python
#seaborn.heatmap(data, *, vmin=None, vmax=None, cmap=None, center=None, annot_kws=None, linewidths=0, linecolor=â€™whiteâ€™, cbar=True, **kwargs)

#Important Parameters:

#data: 2D dataset that can be coerced into an ndarray.
#vmin, vmax: Values to anchor the colormap, otherwise they are inferred from the data and other keyword arguments.
#cmap: The mapping from data values to color space.
#center: The value at which to center the colormap when plotting divergent data.
#annot: If True, write the data value in each cell.
#fmt: String formatting code to use when adding annotations.
#linewidths: Width of the lines that will divide each cell.
#linecolor: Color of the lines that will divide each cell.
#cbar: Whether to draw a colorbar.
#cmap=color of map

sns.heatmap(df.isnull(),cmap='viridis')
```

```python
#replacing null value in age column
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
# Explore survival counts

sns.countplot(x='Survived',data=df)
```

```python
# Explore survival counts based on gender
#You can use the hue parameter when creating pairplots

sns.countplot(x='Survived',hue='Sex',data=df)
plt.show()
```

```python
# Explore survival counts based on passenger class

sns.countplot(x='Survived',hue='Pclass',data=df,palette='rainbow')
```

```python
#distribution of age - whether age follow normal distribution or not.
#help to see that what avg people bore in bot
#your code is generating a simple histogram of the 'Age' column, without the KDE,
#using dark red bars and dividing the data into 50 bins.
# bins=50: This parameter determines the number of bins (or intervals) into which
# the data will be divided. In this case, it sets the number of bins to 50, indicating that 
# the histogram will display the distribution of age values across 50 intervals.

sns.displot(df['Age'],color='darkred',bins=50)
```

```python
# Distribution of ticket fares
plt.figure(figsize=(8, 4))
sns.histplot(df['Fare'],bins=40,color='blue')
plt.title('Distribution of Ticket Prices (Fare)')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
