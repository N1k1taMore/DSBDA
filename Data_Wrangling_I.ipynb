```python
# DATA NORMLIZATION
# Normalization is the process of transforming values of several variables into a similar range. Typical normalizations include scaling the variable so the variable average is 0, scaling the variable so the variance is 1, or scaling variable so the variable values range from 0 to 1
```

```python
#MIN-MAX SCALER
print('horsepower column before scaling')
df['horsepower']
```

```python
print('horse power column after scaling')
scaler=MinMaxScaler()
normalized_data =scaler.fit_transform(df['horsepower'].values.reshape(-1,1))
print(normalized_data)
```

```python
#manullay without using sklearn
df['horsepower']=(df['horsepower']-df['horsepower'].min())/(df['horsepower'].max()-df['horsepower'].min())
print("AFTER MIN_MAX SCALING")
df['horsepower'].head()
```

```python
# Z-SCORE SCALING // also called as Standadization
#with library - skleran (stanadrdScaler)
scaler = StandardScaler()

print("AFTER STANDARDIZATION - enginesize")
df['enginesize'] = scaler.fit_transform(df[['enginesize']])
df['enginesize'].head()
```

```python
#manullay without using sklearn
df['enginesize'] = (df['enginesize'] - df['enginesize'].mean() )/df['enginesize'].std()
df['enginesize'].head()
```

```python
# DECIMAL SCALING
print("BEFORE DECIMAL SCALING - citympg	")
df['citympg'].head()
```

```python
print("AFTER DECIMAL SCALING - citympg	")
df['citympg'] = df['citympg']/10
df['citympg'].head()
```

```python
# DATATYPE CONVERSION - Turn categorical variables into quantitative variables
#Implicite conversion - astype()
print("BEFORE CONVERSION - curbweight")
df['curbweight'].dtype
```

```python
print("AFTER CONVERSION - df['curbweight'].dtype")
df['curbweight'] = df['curbweight'].astype(float)
df['curbweight'].dtype
```

```python
# CATEGORIAL VALUES INTO NUMERICAL VALUES 
# Categorical values, also known as categorical variables or qualitative variables, 
#represent data that falls into specific categories or groups.
# For example, colors (red, blue, green) 
```

```python
unique_val = df['carbody'].unique()

print("The unique categorical values in the column 'carbody' are:")
print(unique_val)
```

```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()

encoded = label.fit_transform(df['carbody'])

new_df = pd.DataFrame({
    'original_' + 'encoded' : df['carbody'],
    'encoded_' + 'carbody' : encoded
})

print(new_df)
```

```python
#using One-Hot-Encoding
#bascially represent unique features in binary format for 
#ex: for red - 00 for green - 01 for blue - 10 for black- 11
```

```python
unique_values = df['aspiration'].unique()

print("The unique categorical values in the column 'aspiration' are:")
print(unique_values)
```

```python
#One-hot encoding converts categorical variables into a binary matrix representation, where each category becomes a new binary column. 
# Assuming df['aspiration'] is a categorical variable
one_hot_encoded = pd.get_dummies(df['aspiration'], prefix='aspiration')

#The [:, 0] syntax specifies that we want to select all rows (:) and the column at index 0. 
# Selecting one column from the one-hot encoded DataFrame
encoded_column = one_hot_encoded.iloc[:, 0]

# Creating the comparison DataFrame
comparison_df = pd.DataFrame({
    'Original_' + 'aspiration': df['aspiration'],
    'Encoded_' + 'aspiration': encoded_column
})

print(comparison_df)
```
